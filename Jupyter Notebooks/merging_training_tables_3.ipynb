```python
import pandas as pd
```

```python
database_type = "/training"
```

```python
merge2 = pd.read_csv('../eICU' + database_type + '/merge2.csv')
merge2
```

```python
diagnosis = pd.read_csv('../eICU' + database_type + '/diagnosis.csv')
diagnosis = diagnosis.astype({'diagnosis': int})
diagnosis
```

```python
finalMerge = pd.merge(merge2, diagnosis,  how='outer', left_on=['patientunitstayid','observationoffset'], right_on = ['patientunitstayid', 'diagnosisoffset'])
del merge2, diagnosis
```

```python
path = "../eICU/training/"
finalMerge.to_csv(path + "merge3.csv", sep=',', index=False, encoding='utf-8')
```
