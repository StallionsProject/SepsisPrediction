```python
import pandas as pd
```

```python
databasePath = "../eICU/training/"
exportPath = "../eICU/training/"
```

#### Reading in Training Tables


```python
lab = pd.read_csv(databasePath + 'lab.csv')
lab
```

```python
vitalperiodic = pd.read_csv(databasePath + '/vitalPeriodic.csv')
vitalperiodic
```

```python
# The original dataframes were deleted to free up resources.  
merge = pd.merge(vitalperiodic, lab,  how='outer', left_on=['patientunitstayid','observationoffset'], right_on = ['patientunitstayid','labresultoffset'])
del lab, vitalperiodic
merge 
```

```python
intakeOutputUrine = pd.read_csv(databasePath + '/intakeOutputUrine.csv')
intakeOutputUrine
```

```python
merge1 = pd.merge(merge, intakeOutputUrine,  how='outer', left_on=['patientunitstayid','observationoffset'], right_on = ['patientunitstayid','intakeoutputoffset'])
del merge, intakeOutputUrine
merge1
```

```python
merge1['observationoffset'] = merge1.apply(lambda x: x['observationoffset'] if pd.notna(x['observationoffset']) else (x['labresultoffset'] if pd.notna(x['labresultoffset']) else x['intakeoutputoffset']), axis = 1)
merge1 = merge1.drop(['labresultoffset', 'intakeoutputoffset'], axis = 1)
merge1
```

```python
merge1.to_csv(exportPath + "merge1.csv", sep=',', index=False, encoding='utf-8')
```

```python
databasePath = "../eICU/training/"
exportPath = "../eICU/training/"
```

```python
merge1 = pd.read_csv(databasePath + 'merge1.csv')
```

```python
diagnosis = pd.read_csv(databasePath + 'diagnosis.csv')
diagnosis = diagnosis.astype({'diagnosis': int})
diagnosis
```

```python
finalMerge = pd.merge(merge1, diagnosis,  how='outer', left_on=['patientunitstayid','observationoffset'], right_on = ['patientunitstayid', 'diagnosisoffset'])
del merge1, diagnosis
finalMerge['observationoffset'] = finalMerge.apply(lambda x: x['observationoffset'] if pd.notna(x['observationoffset']) else x['diagnosisoffset'], axis = 1)
finalMerge = finalMerge.drop(['diagnosisoffset'], axis = 1)
finalMerge.sort_values(['patientunitstayid','observationoffset'])
finalMerge
```

```python
finalMerge.to_csv(exportPath + "finalMerge.csv", sep=',', index=False, encoding='utf-8')
```
