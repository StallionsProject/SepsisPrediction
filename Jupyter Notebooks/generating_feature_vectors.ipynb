```python
import pandas as pd
```

```python
database_type = "/training"
```

```python
cleanedData = pd.read_csv('../eICU' + database_type + '/cleanedData.csv')
cleanedData
```

```python
cleanedData = cleanedData.astype({'diagnosis': int})
cleanedData
```

```python
idList = sorted(set(list(cleanedData['patientunitstayid'])))
```

```python
cols = ['patientid', 'temperature_min', 'temperature_max', 'temperature_mean', 'temperature_std', 'temperature_kurtosis', 'temperature_skew', 
        'heartrate_min', 'heartrate_max', 'heartrate_mean', 'heartrate_std', 'heartrate_kurtosis', 'heartrate_skew', 'respiration_min', 
        'respiration_max', 'respiration_mean', 'respiration_std', 'respiration_kurtosis', 'respiration_skew', 'systemicsystolic_min', 
        'systemicsystolic_max', 'systemicsystolic_mean', 'systemicsystolic_std', 'systemicsystolic_kurtosis', 'systemicsystolic_skew', 
        'wbcx1000_min', 'wbcx1000_max', 'wbcx1000_mean', 'wbcx1000_std', 'wbcx1000_kurtosis', 'wbcx1000_skew', 'lactate_min', 'lactate_max', 
        'lactate_mean', 'lactate_std', 'lactate_kurtosis', 'lactate_skew', 'creatinine_min', 'creatinine_max', 'creatinine_mean', 'creatinine_std', 
        'creatinine_kurtosis', 'creatinine_skew', 'urineoutputbyweight_min', 'urineoutputbyweight_max', 'urineoutputbyweight_mean', 
        'urineoutputbyweight_std', 'urineoutputbyweight_kurtosis', 'urineoutputbyweight_skew', 'diagnosis']
# cleanedData[cols].isna().sum().sum()
```

```python
features = pd.DataFrame(columns=cols)
x = 0
for id in idList:
    df = cleanedData[cleanedData['patientunitstayid'] == id]
    lastPos = len(df.index)-1
    sepsisCount = 0
    
    if x%5000 ==0:
        print(x)
    x+=1
    
    while lastPos > 0:
        if df.iloc[lastPos]['diagnosis'] == 1:
            sepsisCount+=1
            sepsisTime = df.iloc[lastPos]['observationoffset']
            
            instance = df[df['observationoffset'] <= sepsisTime]
            features = append_features(features, instance, 1)
            
        lastPos-=1
        
    if sepsisCount == 0:
        features = append_features(features, df, 0)

features
```
