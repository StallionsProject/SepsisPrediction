```python
%%time
x = 0
for id in idList[:15000]:
    df = cleanedData[cleanedData['patientunitstayid'] == id]
    lastPos = len(df.index)-1
    sepsisCount = 0
    
    if x%5000 ==0:
        print(x)
    x+=1
    
    while lastPos > 0:
        if df.iloc[lastPos]['diagnosis'] == 1:
            sepsisCount+=1
            sepsisTime = df.iloc[lastPos]['observationoffset']
            
            instance = df[df['observationoffset'] <= sepsisTime]
            append_features(features, instance, 1, sepsisCount)
            
        lastPos-=1
        
    if sepsisCount == 0:
        append_features(features, df, 0, sepsisCount)

df = pd.DataFrame(features)
df
```
