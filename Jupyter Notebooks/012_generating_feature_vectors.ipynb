```python
def generate_features(idList):
    start_time = time.time()

    x = 0
    for id in idList:
        df = cleanedData[cleanedData['patientunitstayid'] == id]
    #     dfExp = exPData[exPData['patientunitstayid'] == id]

        lastPos = len(df.index)-1
        sepsisCount = 0

        if x%5000 ==0:
            print('Id Number: ', x, " --- %s minutes  ---" % ((time.time() - start_time) / 60))
        x+=1

        while lastPos > 1:
            if df.iloc[lastPos]['diagnosis'] == 1 and df.iloc[lastPos-1]['diagnosis'] == 0:
                sepsisCount+=1
                sepsisTime = df.iloc[lastPos]['observationoffset']

                instance = df[df['observationoffset'] <= sepsisTime]
                append_features(id, features, instance, df.iloc[lastPos]['diagnosis'], sepsisCount)

            lastPos-=1

        if sepsisCount == 0:
            append_features(id, features, df, df.iloc[lastPos]['diagnosis'], sepsisCount)


    return pd.DataFrame(features)
```
